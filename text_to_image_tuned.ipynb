```python
import torch
from torch import autocast
from diffusers import StableDiffusionPipeline
```

```python
# Load the pre-trained Stable Diffusion model
model_id = "CompVis/stable-diffusion-v1-4"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")  # Move model to GPU for faster inference
```

```python

def generate_image_from_description(description):
    # Generate the image
    with autocast("cuda"):
        image = pipe(description).images[0]

    # Save and show the image
    image.save("generated_image.png")
    image.show()
```

```python
# Example usage
description = input("Enter the description of the image: ")
generate_image_from_description(description)
```

```python
# Experiment with hyperparameters
guidance_scale = 18.0
num_inference_steps = 100
generator = torch.manual_seed(42)
```

```python
# Generate the image
prompt = "A programmed humanoid robot having screen teaching the primary school students in india"
with autocast("cuda"):
    generated_image = pipe(prompt, guidance_scale=guidance_scale, num_inference_steps=num_inference_steps, generator=generator).images[0]

# Save and display the image
generated_image.save("enhanced_image.png")
generated_image.show()
```
